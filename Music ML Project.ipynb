Simple Machine Learning project to build a model that can predict a persons preferred music genre based on age and gender

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
import joblib
```

```python
df = pd.read_csv('music.csv')
df
```

```python
# input dataset
X = df.drop(columns=['genre'])
X
```

```python
# output dataset
y = df['genre']
y
```

```python
# By convention, we use around 80% of the data to train the model and 20% to verify that the model is correct

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

predictions = model.predict(X_test)
predictions
```

```python
X_test
```

```python
# If the prediction and score is ran multiple times, we will see the accuracy of the model vary from 0.5 to 1.0

score = accuracy_score(y_test, predictions)
score
```

```python
# Saving a trained model
joblib.dump(model, 'music.recommender.joblib')
```

```python
# Loading a trained model
model = joblib.load('music.recommender.joblib')
```

```python
tree.export_graphviz(model, out_file='music-recommender.dot',
                     feature_names=['age', 'gender'],
                     class_names=sorted(y.unique()), 
                                        label='all', rounded=True, filled=True)
```
